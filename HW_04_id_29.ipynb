```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Generate two sets of random data from normal distributions
np.random.seed(42)

# Sample 1
mean1 = 10
std_dev1 = 2
sample1 = np.random.normal(mean1, std_dev1, 50)

# Sample 2
mean2 = 12
std_dev2 = 3
sample2 = np.random.normal(mean2, std_dev2, 50)

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=False) # Assuming unequal variances

# Print the results
print(f"t-statistic: {t_statistic:.2f}")
print(f"p-value: {p_value:.6f}")

# Conclusion
if p_value < 0.05:
    print("The means of the two samples are significantly different.")
else:
    print("There is no significant difference between the means of the two samples.")

# Visualization
plt.figure(figsize=(6, 4))
plt.bar(['Sample 1', 'Sample 2'], [np.mean(sample1), np.mean(sample2)], color=['blue', 'red'], alpha=0.7)
plt.errorbar(['Sample 1', 'Sample 2'], [np.mean(sample1), np.mean(sample2)], yerr=[np.std(sample1, ddof=1), np.std(sample2, ddof=1)], fmt='k_', capsize=68)
plt.title('Comparison of Sample Means')
plt.ylabel('Mean Value')
plt.grid(True)
plt.show()
```
